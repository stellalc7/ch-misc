```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
```

```python
d=np.array([6,5,4,3])            # mm
r=d/20                           # cm
area=np.pi*r**2
flow=500/60                      # cm3/s
velocity=0.01*flow/area          # m/s
```

```python
df=pd.DataFrame(np.column_stack((d,r,area,velocity)))
df=df.rename(columns={0:'diameter (mm)',1:'radius (cm)',
                      2:'area (cm2)',3:'velocity (m/s)'})
```

```python
plt.figure(figsize=(9,6))
plt.plot(d,velocity,'*-')
plt.xlim([2,7])
plt.ylim([0,1.4])
plt.xlabel('Diameter (mm)')
plt.ylabel('Velocity (m/s)')
plt.title('Diameter v. Velocity')
```
